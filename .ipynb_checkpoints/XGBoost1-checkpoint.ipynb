```python
import xgboost as xgb #install as "pip install xgboost"
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("Training.csv")
```

```python
#y contins a categorical variable, we encode it using label encoder
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(dataset['prognosis'])
```

```python
# using describe() to give summary statistics of the continous but not 
#categorical variables
dataset.describe()
```

```python
#split the x and y dataset
x = dataset.iloc[:,:-1]
y = label_encoder.fit_transform(dataset['prognosis'])
```

```python
# we convert the dataset into an optimized data structure that XGBoost supports
#and enhances the performance and efficiency
data_dmatrix = xgb.DMatrix(data=x, label = y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=37)
```

```python
#learning rate = range[0,1]
#regularization parameter (alpha = large values lead to more regularization)
#n_estimators = number of trees you want to build
#objective=determines the loss function to be used like reg:linear for regression problems, reg:logistic for classification problems with only decision,
#binary:logistic for classification problems with probability.
xgb_clf = xgb.XGBClassifier(objective ='reg:logistic', learning_rate = 0.1,
                max_depth = 15, alpha = 10, n_estimators = 10)
```

```python
xgb_clf.fit(X_train,y_train)
y_pred= xgb_clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
