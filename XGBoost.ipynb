```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
```

```python
#split the x and y dataset
x = dataset.iloc[:,:-1]
y = label_encoder.fit_transform(dataset['prognosis'])
```

```python
x.dtypes
```

```python
# we convert the dataset into an optimized data structure that XGBoost supports
#and enhances the performance and efficiency
data_dmatrix = xgb.DMatrix(data=x, label = y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
```

```python
xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10,use_label_encoder=True)
```

```python
xg_reg.fit(X_train,y_train)

preds = xg_reg.predict(X_test)
```

```python
score = xg_reg.score(X_test,y_test)
```
